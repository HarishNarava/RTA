```python
# Imputing nulls of all the columns
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np



# Encode categorical variables
label_encoders = {}
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        # Fill missing values with a placeholder
        df[column] = df[column].fillna('Missing')
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le

# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=25)

# Impute missing values
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Decode the categorical variables back
for column in label_encoders:
    df_imputed[column] = label_encoders[column].inverse_transform(df_imputed[column].astype(int))

df_imputed.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer

# Initialize the KNN Imputer
imputer = KNNImputer(n_neighbors=50)

# First, we convert the object columns to 'category' type
for column in df_imputed.columns:
    if df_imputed[column].dtype == 'object':
        df_imputed[column] = df_imputed[column].astype('category')

# Save the category codes and category names for later use
category_codes = {column: df_imputed[column].cat.codes for column in df_imputed.columns if df_imputed[column].dtype.name == 'category'}
category_names = {column: df_imputed[column].cat.categories for column in df_imputed.columns if df_imputed[column].dtype.name == 'category'}

# Encoding the categorical columns before applying KNN imputer
encoded_df = df_imputed.copy()
for column in df_imputed.columns:
    if df_imputed[column].dtype.name == 'category':
        encoded_df[column] = df_imputed[column].cat.codes

# Apply KNN imputer
df_imputed = pd.DataFrame(imputer.fit_transform(encoded_df), columns=df_imputed.columns)

# Convert the imputed values back to categorical data
for column in df_imputed.columns:
    if column in category_codes:
        df_imputed[column] = pd.Categorical.from_codes(df_imputed[column].round().astype(int), category_names[column])

# Check if there are any remaining NaN values
if df_imputed.isnull().sum().sum() > 0:
    # For numerical columns, you can fill NaNs with column means
    for column in df_imputed.select_dtypes(include=[np.number]).columns:
        df_imputed[column].fillna(df_imputed[column].mean(), inplace=True)

    # For categorical columns, you can fill NaNs with the most frequent category
    for column in df_imputed.select_dtypes(include=['category']).columns:
        df_imputed[column].fillna(df_imputed[column].mode()[0], inplace=True)

df_imputed.head()
```

```python
#4th obs -  Encoding 
def label_encoder(df, feats): 
    for feat in feats:    
        feat_val = list(np.arange(df[feat].nunique()))
        feat_key = list(df[feat].sort_values().unique())
        feat_dict = dict(zip(feat_key, feat_val))
        df[feat] = df[feat].map(feat_dict)
    return df

df_imputed = label_encoder(df_imputed, df_imputed.drop(['Accident_severity'], axis=1).columns)
df_imputed.shape
```

```python
from collections import Counter
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Function to calculate and print class distribution
def print_class_distribution(y, label):
    counter = Counter(y)
    print("=============================")
    print(f"Class distribution ({label}):")
    for k, v in counter.items():
        per = 100 * v / len(y)
        print(f"Class= {k}, n={v} ({per:.2f}%)")

# Print class distribution before SMOTE
print_class_distribution(y_train, "Before SMOTE")

# Apply SMOTE
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)

# Print class distribution after SMOTE
print_class_distribution(y_train, "After SMOTE")

print("=============================")
print(f"Upsampled data shape: {X_train.shape}, {y_train.shape}")
```

```python
# BASE MODELLING
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier


def modelling(X_train, y_train, X_test, y_test, **kwargs):
    scores = {}
    models = []
    if 'xgb' in kwargs.keys() and kwargs['xgb']:
        xgb = XGBClassifier()
        xgb.fit(X_train._get_numeric_data(), np.ravel(y_train, order='C'))
        y_pred = xgb.predict(X_test._get_numeric_data())
        scores['xgb']= [accuracy_score(y_test, y_pred)]
        models.append(xgb)

    if 'rf' in kwargs.keys() and kwargs['rf']:
        rf = RandomForestClassifier(n_estimators=200)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_test)
        scores['rf']= [accuracy_score(y_test, y_pred)]
        models.append(rf)

    if 'extree' in kwargs.keys() and kwargs['extree']:
        extree = ExtraTreesClassifier()
        extree.fit(X_train, y_train)
        y_pred = extree.predict(X_test)
        scores['extree'] = [accuracy_score(y_test, y_pred)]
        models.append(extree)

    if 'svc' in kwargs.keys() and kwargs['svc']:
        svc = SVC()
        svc.fit(X_train, y_train)
        y_pred = svc.predict(X_test)
        scores['svc'] = [accuracy_score(y_test, y_pred)]
        models.append(svc)

    if 'knn' in kwargs.keys() and kwargs['knn']:
        knn = KNeighborsClassifier()
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        scores['knn'] = [accuracy_score(y_test, y_pred)]
        models.append(knn)

    if 'lgbm' in kwargs.keys() and kwargs['lgbm']:
        lgbm = LGBMClassifier(verbose=-1)
        lgbm.fit(X_train, y_train)
        y_pred = lgbm.predict(X_test)
        scores['lgbm'] = [accuracy_score(y_test, y_pred)]
        models.append(lgbm)


        
    return scores, models
```

```python
scores,models = modelling(X_train,y_train, X_test, y_test, xgb=True, rf=True, extree=True,svc=True,knn=True,lgbm=True)
print(scores)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import ExtraTreesClassifier
import joblib

selected_features=['Minutes', 'Day_of_week', 'Number_of_vehicles_involved', 'Number_of_casualties', 'Light_conditions', 'Driving_experience', 'Road_surface_conditions', 'Types_of_Junction', 'Hour', 'Type_of_collision']

X_subset = X[selected_features]
y = df_imputed['Accident_severity']

# Apply SMOTE
smote = SMOTE()
X_upsampled, y_upsampled = smote.fit_resample(X_subset, y)

# Split the upsampled data
X_train, X_test, y_train, y_test = train_test_split(X_upsampled, y_upsampled, test_size=0.2, random_state=42)

# Initialize the ExtraTreesClassifier with the best parameters
model = ExtraTreesClassifier(ccp_alpha= 0.0, 
                                    criterion= "entropy", 
                                    max_depth= 15, 
                                 
                                    min_samples_leaf= 1,
                                    min_samples_split= 2, 
                                    n_estimators= 300,
                                    class_weight = 'balanced'
                                    )
model.fit(X_train, y_train)
```

```python

from sklearn.metrics import classification_report

# Predict the test set results
y_pred = model.predict(X_test)

# Generate the classification report
report = classification_report(y_test, y_pred)

print(report)
```
